```python
datos['CVE_CLPV'] = datos['CVE_CLPV'].astype(float)
```

```python
## Para pasar de str a int
datos.CVE_CLPV = datos.CVE_CLPV.replace({'O':'0'}, regex=False)
datos.CVE_CLPV = datos.CVE_CLPV.replace({'6781|':'6781'}, regex=False)
datos.CVE_CLPV = datos.CVE_CLPV.replace({'3612C':'3612'}, regex=False)
datos.CVE_CLPV = datos.CVE_CLPV.replace({'      ':'0'}, regex=False)
datos.CVE_CLPV = datos.CVE_CLPV.replace({'2385A':'2385'}, regex=False)
datos.CVE_CLPV = datos.CVE_CLPV.replace({'2510A':'2510'}, regex=False)
datos.CVE_CLPV = datos.CVE_CLPV.replace({'8440Q':'8440'}, regex=False)
datos.CVE_CLPV = datos.CVE_CLPV.replace({'M':'0'}, regex=False)
#Se tiene que ir uno por uno (error)
```

```python
f1 = datos[(datos['CVE_CLPV']>= 1000) & (datos['CVE_CLPV']<= 2000)]
f1
```

```python
#Analizar agrupaciones de una columnas
datos2 = f2.groupby(['CVE_VEND'])['CVE_VEND'].count()
datos2
```
